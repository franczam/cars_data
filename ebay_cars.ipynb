```python
autos['brand'].value_counts(normalize = True)
```

```python
# German manufacturers make up top 5 brands. 
# Limit analysis to brands that make up more than 5% of total sales data. 
```

```python
brands = autos['brand'].value_counts(normalize = True)
most_common_brands = brands[brands > 0.05].index
most_common_brands
```

```python
brands_mean_price = {}

for brands in most_common_brands:
    brand_only = autos[autos['brand'] == brands]
    mean_price = brand_only['price'].mean()
    brands_mean_price[brands] = int(mean_price)
    
brands_mean_price
```

```python
# Price gap bewteen top 5 brands. Audi, BMW, and Mercedes Benz prices higher. Opel is the least expensive. Volkswagen is in the middle (probably explaining it's popularity)
```
